MLxtend kütüphanesi, market sepeti analizi ve Apriori algoritmasının uygulanması için kullanılan yaygın bir Python kütüphanesidir. Bu kütüphane, market sepeti analizinde kullanılan Apriori algoritması için fonksiyonlar sağlar.

MLxtend kütüphanesi, Apriori algoritması için birkaç farklı çıktı verir. Bu çıktılar arasında şunlar bulunur:

Frequent Itemsets: Bu, belirli bir minimum destek seviyesine sahip olan öğelerin bir arada görünme sıklıklarını gösteren bir çıktıdır. Yani, bu çıktı, belirli bir destek seviyesinin üzerindeki öğelerin bir arada ne sıklıkta görüldüğünü gösterir.

Association Rules: Bu çıktı, Apriori algoritmasının asıl amacı olan bir öğenin varlığı durumunda diğer öğelerin ne sıklıkla görüldüğünü gösterir. Bu çıktı, "destek" ve "güven" kavramlarına dayanır. Destek, bir ürün kombinasyonunun genel olarak ne kadar sıklıkta görüldüğünü gösterirken, güven, belirli bir ürün kombinasyonunun içindeki bir ürünün ne sıklıkla görüldüğünü gösterir.

Öteki İstatistiksel Özellikler: Bu özellikler arasında destek, güven ve lift gibi kavramlar yer alır. Destek, bir kuralın ne kadar sıklıkla görüldüğünü gösterirken, güven, bir ürün kombinasyonunun içindeki bir ürünün ne sıklıkla görüldüğünü gösterir. Lift ise, iki ürünün birlikte satın alınma olasılığının, bağımsız olarak satın alma olasılıklarına göre ne kadar arttığını gösterir.

Bu çıktılar, işletmelerin müşterilerin davranışlarını anlamalarına ve ürünlerini müşteri ihtiyaçlarına göre düzenlemelerine yardımcı olabilir. Bu sayede, müşterilerin ihtiyaçlarını daha iyi karşılayan ve daha fazla satış yapabilen bir işletme olabilirsiniz.

PMDARIMA kütüphanesi, otomatik zaman serisi modelleme için kullanılan bir Python kütüphanesidir. Bu kütüphane, Auto ARIMA ve SARIMAX algoritmaları için fonksiyonlar sağlar.

Auto ARIMA algoritması, zaman serileri verilerindeki trend, mevsimsellik ve gürültüyü otomatik olarak belirlemek için kullanılan bir makine öğrenimi algoritmasıdır. Bu algoritma, en iyi modeli belirlemek için birden fazla iterasyon yapar ve belirli bir performans ölçütünü kullanarak en iyi modeli seçer. Auto ARIMA algoritması, model oluşturma sürecinde kullanılan parametrelerin (örneğin, ARIMA(p,d,q) parametreleri) otomatik olarak belirlenmesi sayesinde modelleme sürecini kolaylaştırır.

SARIMAX (Seasonal AutoRegressive Integrated Moving Average with eXogenous regressors) algoritması ise, zaman serileri verilerindeki trend, mevsimsellik ve gürültüyü modellemek için kullanılan bir makine öğrenimi algoritmasıdır. Bu algoritma, Auto ARIMA algoritması gibi, bir zaman serisi verisi üzerinde belirli bir performans ölçütüne dayalı olarak en iyi modeli otomatik olarak seçer. Ancak SARIMAX algoritması, Auto ARIMA algoritmasından farklı olarak, exogenous değişkenleri de dahil edebilir. Exogenous değişkenler, zaman serisindeki değişkenlerin yanı sıra dışsal faktörlerin de modelleme sürecine dahil edilmesine olanak tanır.

Auto ARIMA algoritmasının çıktıları arasında, en iyi modelin parametreleri (p, d, q), mevsimsel parametreler (P, D, Q, s), en düşük AIC, BIC ve HQIC değerleri gibi bilgiler yer alır.

SARIMAX algoritması ise, otomatik olarak seçilen en iyi modelin parametreleri (p, d, q, P, D, Q, s), modelin tahminleri, modelin öğrenme sürecindeki hatalar, modelin doğruluğunu ölçmek için kullanılan metrikler ve tahmin aralıkları gibi bilgileri sağlar.

Bu çıktılar, işletmelerin geçmiş verileri analiz ederek gelecekteki satışları tahmin etmelerine ve iş kararlarını buna göre planlamalarına yardımcı olabilir.














# MARKET BASKET ANALYSIS

```python
import pandas as pd

df = pd.read_excel("market_sales_data.xlsx")
df.head(3)
```

```python
# AYNI FATURAYA AİT SATIŞLARI GRUPLA VE LİSTEYE ÇEVİR
sales_data = df[["InvoiceID","PRODUCT_NAME"]].groupby('InvoiceID')["PRODUCT_NAME"].apply(list).reset_index()
sales_data.tail(8)
```

```python
# MARKET SEPET ANALİZİ İÇİN GEREKLİ İMPORTLAR

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules


# TRANSACTION ENCODER -> Faturaları ürün içerme durumuna göre matrix haline getirir.
te = TransactionEncoder()
one_hot = te.fit_transform(sales_data["PRODUCT_NAME"])
one_hot_df = pd.DataFrame(one_hot, columns=te.columns_)

one_hot_df
```

```python
# DATAYI APRIORI ALGORTMASI İLE BUDA (PRUNING)
freq_itemsets = apriori(one_hot_df, min_support=0.0001, use_colnames=True)

# BUDANMIŞ DATADAN ALAKA METRİKLERİNİ HESAPLA
rules = association_rules(freq_itemsets, metric='lift', min_threshold=1)

rules["antecedents"] = rules["antecedents"].apply(lambda x: ",".join(x))
rules["consequents"] = rules["consequents"].apply(lambda x: ", ".join(x))

rules
```

```python
import matplotlib.pyplot as plt 


# EN YÜKSEK 25 LİFT DEĞERİNE SAHİP ÇİFTLERİ AL
top_rules = rules.sort_values("lift", ascending=False).head(25)

plt.figure(figsize=(12,8))

# GÖRSELDEKİ BARLARI AYIRT EDEBİLMEK İÇİN HER BİRİNE FARKLI BİR RENK ATA

colors = ["#EC7462","#C49CCA","#25BDBC","#15A4C2","#A1B6DF"]

#LIST COMPREHENSION
color = [colors[i % len(colors)] for i in range(len(top_rules))]


# YATAY -HORIZONTAL BAR PLOT 
plt.barh(range(len(top_rules)), top_rules['lift'],color = color)
plt.yticks(range(len(top_rules)), top_rules['antecedents'] + ' -> ' + top_rules['consequents'])
plt.ylabel('Lift')
plt.title('Antecedent-Consequent Lift Values')

plt.grid()
plt.tight_layout()

plt.show()
```

```python
test_months  = 18

train = monthly_sales[:-test_months] # son 18 aya kadar
test = monthly_sales[-test_months:] # son 18 ay

model = auto_arima(y = train,
                   #X = train_exog, # eğer dış etmen eklemek isteseydik (REKLAM KAMPANYASI - EKONOMİK KRIZ - KÜRESEL ISINMA)
                   m = 12,
                   seasonal = True,
                   stepwise = False)
```


R-squared (R²), bir regresyon modelinin uygunluğunu ölçmek için kullanılan bir istatistiksel ölçüttür. R² değeri, bağımsız değişkenlerin bağımlı değişkendeki değişkenliği açıklama oranını ifade eder. Bu değer, 0 ile 1 arasında bir değer alır ve ne kadar yüksek olursa, modelin verileri ne kadar iyi açıkladığına dair o kadar iyi bir gösterge olur. R² değeri 1'e yaklaştıkça, model verileri daha iyi açıklar.

Öte yandan, Ortalama Yüzde Hata Oranı (MAPE), bir tahmin modelinin doğruluğunu ölçmek için kullanılan bir ölçüttür. MAPE, gerçek değerlerle tahmin edilen değerler arasındaki farkı, gerçek değerlerin yüzdesi olarak ifade eder. MAPE değeri genellikle yüzde olarak ifade edilir ve ne kadar düşük olursa, modelin tahminleri o kadar doğru olur.

Her iki ölçüt de farklı amaçlar için kullanılır. R², regresyon modellerinin performansını değerlendirmek için kullanılırken, MAPE tahmin modellerinin performansını değerlendirmek için kullanılır.
